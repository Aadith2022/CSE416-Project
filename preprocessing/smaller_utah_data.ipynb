```python
import geopandas as gpd

# Load your large GeoJSON file
gdf = gpd.read_file("./clean_data/texas_data/texas_precinct_data.geojson")
gdf.plot()
# Simplify the geometries with a specified tolerance (e.g., 0.001)
# The tolerance is the maximum distance between the original geometry and the simplified version
gdf['geometry'] = gdf['geometry'].simplify(tolerance=0.005, preserve_topology=True)

# Save the simplified GeoDataFrame as a new GeoJSON file
gdf.to_file("texas_simplified_precinct_data.geojson", driver="GeoJSON")
```
