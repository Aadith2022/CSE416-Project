```python
import geopandas as gpd

# Load your large GeoJSON file
gdf = gpd.read_file("utah_current_district_data.geojson")
gdf.plot()
# Simplify the geometries with a specified tolerance (e.g., 0.001)
# The tolerance is the maximum distance between the original geometry and the simplified version
gdf['geometry'] = gdf['geometry'].simplify(tolerance=0.001, preserve_topology=True)

# Save the simplified GeoDataFrame as a new GeoJSON file
gdf.to_file("utah_simplified_current_district_data.geojson", driver="GeoJSON")
```
