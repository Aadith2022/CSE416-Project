```python
import geopandas as gpd
```

## Utah Precinct Cleaning Of Data

```python
file_path = "./utah_data/aggregated_pre.geojson"

gdf = gpd.read_file(file_path)

print(gdf.head())

print(gdf.crs)
```

```python
unwanted_columns = [
    "vistapre",
    "G20PRELJOR",
    "G20PREGHAW",
    "G20PRECBLA",
    "G20PREIPIE",
    "G20PREIWES",
    "G20PREIMCH",
    "G20PREILAR",
    "G20PREOWRI",
    "G20GOVRCOX",
    "G20GOVDPET",
    "G20GOVLCOT",
    "G20GOVADUE",
    "G20GOVOWRI",
    "G20ATGRREY",
    "G20ATGDSKO",
    "G20ATGLBAU",
    "G20AUDRDOU",
    "G20AUDCOST",
    "G20AUDUFAB",
    "G20TRERDAM",
    "G20TRELSPE",
    "G20TRECPRO",
  ]
gdf = gdf.drop(columns=unwanted_columns)
gdf.head()
```

```python
columnNameMapping = {
    "DISTRICT": "index",
    "PP_TOTAL": "Population",
    "resultspre": "Precinct",
    "G20PRERTRU": "Republican",
    "G20PREDBID": "Democrat",
    "PP_WHTALN": "White",
    "PP_BAAALN": "Black",
    "PP_HISPLAT": "Hispanic",
    "PP_ASNALN": "Asian",
    "PP_HPIALN": "Pacific",
    "PP_NAMALN": "Native",
    "PP_OTHALN": "Other",
}

gdf = gdf.rename(columns=columnNameMapping)
gdf.head()
```

```python
unwated_county_id = ["CountyID"]
gdf = gdf.drop(columns=unwated_county_id)
```

```python
gdf.head()
```

```python
gdf.to_file("utah_precinct_data.geojson", driver='GeoJSON')
```

## Utah District Cleaning

```python
file_path = "./utah_data/aggregatedUtahDistricts.geojson"

gdf = gpd.read_file(file_path)

print(gdf.head())

print(gdf.crs)
```

```python
representative_names = ["Blake Moore", "Celeste Maloy", "John Curtis", "Burgess Owens"]
incumbent_party = ["Republican", "Republican", "Republican", "Republican"]
```

```python
new_column_names = ["Representative", "Incumbent"]
gdf[new_column_names[0]] = representative_names
gdf[new_column_names[1]] = incumbent_party
```

```python
gdf["Representative"]
```

```python
columnNameMapping = {
        "DISTRICT": "District",
        "Representative": "Representative",
        "Incumbent": "Incumbent",
        "PP_TOTAL": "Population",
        "G20PRERTRU": "Republican",
        "G20PREDBID": "Democrat",
        "PP_WHTALN": "White",
        "PP_BAAALN": "Black",
        "PP_HISPLAT": "Hispanic",
        "PP_ASNALN": "Asian",
        "PP_HPIALN": "Pacific",
        "PP_NAMALN": "Native",
        "PP_OTHALN": "Other",
}

gdf = gdf.rename(columns=columnNameMapping)
unwanted_columns = ["TAPERSONS", "VAPERSONS"]
gdf = gdf.drop(columns=unwanted_columns)
gdf.head()
```

```python
gdf.to_file("utah_current_district_data.geojson", driver="GeoJSON")
```
